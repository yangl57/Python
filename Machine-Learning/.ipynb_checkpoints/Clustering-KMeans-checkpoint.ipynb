## Clustering: K-Means

K Means is one type of unsupervised clustering. It looks for cluster centers which are the mean of the points within them. Here is a demo of it with sklearn. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
x,y = make_blobs(n_samples=500,centers=5,random_state=2,cluster_std=0.8)
plt.scatter(x[:,0],x[:,1],s=50)
```

```python
from sklearn.cluster import KMeans
est=KMeans(5)
est.fit(x)
y_kmeans = est.predict(x)
plt.scatter(x[:,0],x[:,1],c=y_kmeans,s=50,cmap='jet')
```

```python
from sklearn.datasets import load_digits 
digits=load_digits()
digits.keys()
```

```python
est=KMeans(n_clusters=10)
clusters=est.fit_predict(digits.data)
```

```python
est.cluster_centers_.shape 
```

Here, we have 10 clusters, each of which is in 64 dimensions. 

```python
fig=plt.figure()
for i in range(10):
    ax=fig.add_subplot(2,5,i+1,xticks=[],yticks=[])
    ax.imshow(est.cluster_centers_[i].reshape((8,8)))
```

```python
from scipy.stats import mode

labels = np.zeros_like(clusters)
for i in range(10):
    print 'i:', i
    mask = (clusters ==i)
    labels[mask] = mode(digits.target[mask])[0]
    print mode(digits.target[mask])[0]
labels
```

```python
from sklearn.decomposition import PCA
x = PCA(2).fit_transform(digits.data)

kwargs = dict(cmap=plt.cm.get_cmap('rainbow',10), edgecolor='none',alpha=0.6)
fig,ax=plt.subplots(1,2,figsize=(14,6))
ax[0].scatter(x[:,0],x[:,1],c=labels, **kwargs)
ax[1].set_title('learned cluster labels')

ax[1].scatter(x[:,0],x[:,1],c=digits.target, **kwargs)
ax[1].set_title('learned cluster labels')
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(digits.target, labels)
```
