# Dimensionality Reduction : Principal Component Analysis 

PCA is an unsupervised method  for dimensionality reduction in data. 

```python
from __future__ import print_function, division
import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
import seaborn as sns
%matplotlib inline
```

```python
np.random.seed(1)
x=np.dot(np.random.random(size=(2,2)),np.random.normal(size=(2,400))).T
plt.plot(x[:,0],x[:,1],'o')
plt.axis('equal')
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(x)
print(pca.explained_variance_)
print(pca.components_)
```

```python
style=['-r','-k']
plt.plot(x[:,0],x[:,1],'o',alpha=0.5)
i=0
for length,vector in zip(pca.explained_variance_,pca.components_):
    v=vector*3*np.sqrt(length)
    plt.plot([0,v[0]],[0,v[1]],style[i],lw=3)
    i+=1
plt.axis('equal')
```

The explained variance indicates the importance in directions. As a result, second principal component could be completed ignored without much loss of information. 

What if we only keep 95% of variance?

```python
clf = PCA(0.95)
x_trans = clf.fit_transform(x)
print (x.shape)
print (x_trans.shape)
```

Data is compressed 50%

```python
x_new = clf.inverse_transform(x_trans)
plt.plot(x[:,0],x[:,1],'o',alpha=0.8)
plt.plot(x_new[:,0],x_new[:,1],'or',alpha=0.8)
plt.axis('equal')
```

After truncating 5% of the variance of this dataset and then reprojecting, the most important features of the data are maintained with 50% reduction of the data.

Looking at this plot for a high-dimensional dataset can help you understand the level of redundancy present in multiple observations.
