```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
%matplotlib inline
from sklearn.datasets import make_blobs
import warnings

def fxn():
    warnings.warn("deprecated", DeprecationWarning)

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    fxn()
    

warnings.filterwarnings('ignore')
```

```python
x, y = make_blobs(n_samples=50,centers=2,random_state=0,cluster_std=0.6)
plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap='jet')
```

```python
from sklearn.svm import SVC 
clf = SVC(kernel='linear')
clf.fit(x,y)
```

```python
def plot_svc_decision_function(clf, ax=None):
    
    if ax is None:
        ax=plt.gca()
    x = np.linspace(ax.get_xlim()[0],ax.get_xlim()[1],30)
    y = np.linspace(ax.get_ylim()[0],ax.get_ylim()[1],30)
    
    Y,X = np.meshgrid(y,x)
    P = np.zeros_like(X)
    
    for i, xi in enumerate(x):
        for j, yj in enumerate(y):
            P[i,j]=clf.decision_function([xi,yj])
    
    ax.contour(X,Y,P,colors='k',
              levels=[-1,0,1], alpha=0.5,
              linestyles=['--','-','--'])
```

```python
g=plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap='spring')
g.axes
plot_svc_decision_function(clf,g.axes);
```

```python
g=plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap='spring')
g.axes
plot_svc_decision_function(clf,g.axes);
plt.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],edgecolors='red',facecolor='none',
           s=200)
```

```python
from IPython.html.widgets import interact
```

```python
def plot_svm(N=20):
    x, y = make_blobs(n_samples=200,centers=2,random_state=0,cluster_std=0.6)
    clf.fit(x,y)
    plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap='spring')
    plt.xlim(-1,4)
    plt.ylim(-1,6)
    plot_svc_decision_function(clf)
    plt.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],
               s=200, edgecolors='red',facecolor='none')

interact(plot_svm, N=[10,200], kernel='linear')
```
